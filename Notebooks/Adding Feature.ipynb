### Add product Average Rating

```python
display(bigdf.groupby('product_id').size())
map_pid = {}
avgRating = []
#siapin dict nyipen avg rating
for i in list(set(bigdf.product_id)):
    map_pid[i] = 0
#isi dict dengan avg rating
for i,group in bigdf.groupby('product_id'):
    map_pid[i] = round(group.review_rating.mean(),2)
    if i == 67.0:
        print(i,round(group.review_rating.mean(),2))
for i in range(len(bigdf)):
    sel = bigdf.iloc[i]
    avgRating.append(map_pid[sel.product_id])
bigdf['product_avg_rating'] = avgRating
bigdf.head()
```

```python
bigdf[bigdf['product_id'] == 67].product_avg_rating
```

### Standard Deviation

```python
map_pid = {}
stdRating = []
#siapin dict nyipen avg rating
for i in list(set(bigdf.product_id)):
    map_pid[i] = 0
#isi dict dengan avg rating
for i,group in bigdf.groupby('product_id'):
    map_pid[i] = np.std(group.review_rating)
    if i == 67.0:
        print(i,np.std(group.review_rating))
for i in range(len(bigdf)):
    sel = bigdf.iloc[i]
    stdRating.append(map_pid[sel.product_id])
bigdf['product_std_rating'] = stdRating
bigdf.head()
```

```python
bigdf[bigdf['product_id'] == 67].product_std_rating
```

### Add Diff from avg rating

```python
diffRat = []
for row in bigdf.iterrows():
    diffRat.append(row[1].review_rating-row[1].product_avg_rating)
bigdf['rating_diff'] = diffRat
bigdf
```

## no opinion and word based sentiment

```python
def read_txt(file):
    temp = []
    f = open(file, "r")
    for x in f:
        temp.append(x.strip('\n'))
    return temp
```

```python
id_positive = read_txt('../Dataset/ID-OpinionWords/positive.txt')
id_negative = read_txt('../Dataset/ID-OpinionWords/negative.txt')
en_positive = read_txt('../Dataset/EN-OpinionWords/positive.txt')
en_negative = read_txt('../Dataset/EN-OpinionWords/negative.txt')
len(en_positive),len(en_negative)
```

```python
no_opinions = []
sentiment = []
positive_word = []
negative_word = []
for i in tqdm(range(len(bigdf))):
    res_positive = [word for word in bigdf['review_text_clean'].iloc[i].split() if (word in id_positive) or (word in en_positive)]
    res_negative = [word for word in bigdf['review_text_clean'].iloc[i].split() if (word in id_negative) or (word in en_negative)]
    positive_word.append(len(res_positive))
    negative_word.append(len(res_negative))
    if (len(res_positive) + len(res_negative)) == 0:
        no_opinions.append('yes')
        sentiment.append('neutral')
    else:
        no_opinions.append('no')
        diff = len(res_positive)-len(res_negative)
        if diff > 0:
            sentiment.append('positive')
        elif diff == 0:
            sentiment.append('neutral')
        else:
            sentiment.append('negative')
bigdf['no_opinions'] = no_opinions
bigdf['sentiment_word'] = sentiment
bigdf['positive_word'] = positive_word
bigdf['negative_word'] = negative_word
```
