```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import re
from html import unescape, escape
from nltk.corpus import stopwords
import nltk
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory
stemmer = StemmerFactory().create_stemmer()
stopwords = StopWordRemoverFactory().create_stop_word_remover()
import unicodedata
from nltk.tokenize import word_tokenize
import pickle
import string
from collections import Counter
from tqdm import tqdm
```

```python
df = pd.read_csv('../dataset/top10cat_balanced.tsv',sep='\t')
df.head()
```

```python
singk = pd.read_csv('../dataset/kumpulan_singkatan_dan_kata_dasar.csv',sep=';',header=None)
singk.columns = ['singkatan','kata_dasar']
singk
```

```python
def stemming(text):
    return stemmer.stem(text)

def remove_link(text):
    return re.sub(r'http\S+', '', text)

def make_unicode(input):
    if type(input) != unicode:
        input =  input.decode('utf-8')
    return input

def normalizee(text):
    i = 0
    j = 0
    while True:
        match = ''
        if text[i+1] == text[i]:
            j = i
            match = text[j]
            while j <= len(text)-2:
                if text[j+1] == text[j]:
                    match += text[j]
                    j+=1
                else:
                    break
            if len(match) > 2:
                text = text.replace(match,text[i])
        if i+1 <= len(text)-2:
            i+=1
        else:
            break
    return text

not_allowed_char = ['\uf04c','\u200d','\xa0','<br>','&nbsp;','.']
def filter_char(line):
    for denied in not_allowed_char:
        line = line.replace(denied,' ')
    return line

def remove_stopwords(line):
    return stopwords.remove(line)

def to_formal(line):
    for word in word_tokenize(line):
        if word in singk['singkatan'].values.tolist(): 
            idx = singk['singkatan'].values.tolist().index(word)
#             print(idx)
            line = line.replace(word,str(singk.iloc[idx]['kata_dasar']))
    return line
```

```python
def cleaning_text(text):
    vocab = []
    new_text = []
    for line in tqdm(text,total=len(text)):
        before = len(line)
        new_line = line
        new_line = remove_link(new_line)
        new_line = filter_char(new_line)
        new_line = unescape(new_line)
        new_line = new_line.replace('\\\\\\\\/','/').replace('<br />',' ')
#         new_line = new_line.encode().decode('unicode_escape')
        new_line = unicodedata.normalize("NFKD", new_line)
        new_line = new_line.translate(str.maketrans(' ', ' ', string.punctuation))
        new_line = normalizee(new_line)
        new_line = remove_stopwords(new_line)
        new_line = filter_char(new_line)
        new_line = to_formal(new_line)
#         new_line = stemming(new_line)
        after = len(new_line)
        new_text.append(' '.join(new_line.strip().split()))
        vocab+=word_tokenize(new_line)
    return vocab, new_text
```

```python
vocab, df['review_text_clean'] = cleaning_text(df.review_text)
```
